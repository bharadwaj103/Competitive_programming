```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    # Update operation (add value to index)
    def update(self, index, value):
        while index <= self.n:
            self.tree[index] += value
            index += index & -index

    # Query operation (sum from 1 to index)
    def query(self, index):
        total = 0
        while index > 0:
            total += self.tree[index]
            index -= index & -index
        return total


# ----------- Main Program -----------

# Input
n = int(input("Enter number of days: "))
attendance = list(map(int, input("Enter attendance points: ").split()))

# Create Fenwick Tree
fenwick = FenwickTree(n)

# Build tree
for i in range(n):
    fenwick.update(i + 1, attendance[i])

# Operation 1: Query till Day 4
day = 4
print("Total attendance till Day", day, "=", fenwick.query(day))

# Operation 2: Update Day 3 from 1 → 2
old_value = attendance[2]
new_value = 2
difference = new_value - old_value
attendance[2] = new_value
fenwick.update(3, difference)

# Operation 3: Query again till Day 4
print("After update, total attendance till Day", day, "=", fenwick.query(day))
```

```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    # Add value to index
    def update(self, index, value):
        while index <= self.n:
            self.tree[index] += value
            index += index & -index

    # Get prefix sum from 1 to index
    def query(self, index):
        total = 0
        while index > 0:
            total += self.tree[index]
            index -= index & -index
        return total


# -------- Main Program --------

# Input
n = int(input("Enter number of days: "))
arr = list(map(int, input("Enter daily sales revenue: ").split()))

# Create Fenwick Tree
fenwick = FenwickTree(n)

# Build tree
for i in range(n):
    fenwick.update(i + 1, arr[i])

# Example Operations

# Query 1: Total till Day 6
print("Total sales revenue till Day 6 =", fenwick.query(6))

# Update: Day 5 from 100 → 160
old_value = arr[4]
new_value = 160
difference = new_value - old_value
arr[4] = new_value
fenwick.update(5, difference)

# Query 2: Total till Day 6 after update
print("After update, total sales revenue till Day 6 =", fenwick.query(6))
```
