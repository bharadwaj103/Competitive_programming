```python
# Maximum Bipartite Matching using DFS (Kuhn's Algorithm)

# Read input
n, m, e = map(int, input().split())

# Adjacency list for left set A (1-based indexing)
graph = [[] for _ in range(n + 1)]

for _ in range(e):
    u, v = map(int, input().split())
    graph[u].append(v)

# matchR[v] = which vertex in A is matched to v in B
matchR = [-1] * (m + 1)

def dfs(u, visited):
    for v in graph[u]:
        if not visited[v]:
            visited[v] = True

            # If v is not matched OR we can re-match its current partner
            if matchR[v] == -1 or dfs(matchR[v], visited):
                matchR[v] = u
                return True
    return False

max_matching = 0

for u in range(1, n + 1):
    visited = [False] * (m + 1)
    if dfs(u, visited):
        max_matching += 1

print("Maximum Matching Size =", max_matching)

print("Matching:")
for v in range(1, m + 1):
    if matchR[v] != -1:
        print(matchR[v], v)
```
