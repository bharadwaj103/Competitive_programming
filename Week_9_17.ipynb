```python
def compute_lps(pattern: str) -> list:
    """
    Compute the LPS (Longest Prefix Suffix) array.

    Args:
        pattern (str): The pattern string

    Returns:
        list: LPS array
    """
    lps = [0] * len(pattern)
    length = 0  # length of previous longest prefix suffix
    i = 1

    while i < len(pattern):
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1

    return lps


def kmp_search(text: str, pattern: str) -> list:
    """
    Perform KMP search and return all matching indices.

    Args:
        text (str): The main text
        pattern (str): The pattern to search

    Returns:
        list: List of starting indices where pattern is found
    """
    lps = compute_lps(pattern)
    result = []

    i = 0  # index for text
    j = 0  # index for pattern

    while i < len(text):
        if text[i] == pattern[j]:
            i += 1
            j += 1

        if j == len(pattern):
            result.append(i - j)
            j = lps[j - 1]

        elif i < len(text) and text[i] != pattern[j]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1

    return result


# ------------------ MAIN ------------------
if __name__ == "__main__":
    text = input("Enter post text: ")
    pattern = input("Enter keyword: ")

    matches = kmp_search(text, pattern)

    if matches:
        for idx in matches:
            print(f"Pattern found at index {idx}")
    else:
        print("Pattern not found")
```
